```python
# Helper to install geopandas: https://medium.com/analytics-vidhya/fastest-way-to-install-geopandas-in-jupyter-notebook-on-windows-8f734e11fa2b
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

fpath = r'C:\Users\utilizador\Documents\a_hjoaquim\semestre2_5oano\PSn\ShapeFiles\gadm36_PRT.gpkg'
shapefile = gpd.read_file(fpath)
shapefile['join'] = shapefile['NAME_1'] + shapefile['NAME_2']

# More on dissolve: https://geopandas.readthedocs.io/en/latest/docs/reference/api/geopandas.GeoDataFrame.dissolve.html
# Basically: does a group by for spacial data.
shp = shapefile.dissolve(by='join')
```

```python
# Getting the countys from the shapefile 
munGeo = shp['NAME_2'].to_list()
shp
```

```python
path_municipios = r'C:\Users\utilizador\Documents\a_hjoaquim\semestre2_5oano\PSn\output.xlsx'
df = pd.read_excel(path_municipios, index_col='Municipios')
df
```

```python
# Strip every county in the list, wich had a space in the end.
df.index = [mun.strip() for mun in df.index]
# Removing 'Portugal' from the county list and also removing white spaces that were in the list itself.
listOfCountys = df.drop('Portugal').index.unique().to_list()
```

```python
# Check what's missing in each of the lists

for m in listOfCountys:
    if m not in munGeo:
        print('Falta',m)
        
print('##########')
        
for m in munGeo:
    if m not in listOfCountys:
        print('Falta',m)
        
# Renaming
shp.loc[shp['NAME_2']=='Ponte de Sôr','NAME_2'] = 'Ponte de Sor'
shp.loc[shp['NAME_2']=='Praia da Vitória','NAME_2'] = 'Vila da Praia da Vitória'
shp.loc[((shp['NAME_2']=='Calheta') & (shp['NAME_1'] == 'Azores')),'NAME_2'] = 'Calheta [R.A.A.]'
shp.loc[((shp['NAME_2']=='Calheta') & (shp['NAME_1']=='Madeira')),'NAME_2'] = 'Calheta [R.A.M.]'
shp.loc[((shp['NAME_2']=='Lagoa') & (shp['NAME_1']=='Azores')),'NAME_2'] = 'Lagoa [R.A.A.]'

munGeo = shp['NAME_2'].to_list()
```

```python
# As we can see below, no more differences!

for m in listOfCountys:
    if m not in munGeo:
        print('Falta',m)
        
print('##########')
        
for m in munGeo:
    if m not in listOfCountys:
        print('Falta',m)
```

```python
# Renaming column
shp=shp.rename(columns={'NAME_1':'Distrito'})
# Changing index so it county based again
shp.index=shp['NAME_2']
```

```python
madeiraCountys = shp[shp['Distrito']=='Madeira'].index.to_list()
azoresCountys = shp[shp['Distrito']=='Azores'].index.to_list()
continentCountys = shp[(shp['Distrito']!='Azores') & (shp['Distrito']!='Madeira')].index.to_list()
```

```python
# Check if we haven't lost anything
len(continentCountys) + len(azoresCountys) + len(madeiraCountys)
```

```python
# Adding info that is in the shapefile to the original excel
# Pandas library automatically find the indexes for us, so we don't have to bother
df['Distrito'] = shp['Distrito']
df['geometry'] = shp['geometry']
df
```

```python
# Aging index
for i in df:
    P1 = df['População Residente'] * df['Jovens (%) ']
    P2 = df['População Residente'] * df['Idosos (%) ']
    df['Índice de envelhecimento new'] = ( P1 / P2 )
    
df
```

```python
def printMap(df,area,title):
    cmap=plt.cm.rainbow
    ax = plt.gca()
    for m in area:
        muni=gpd.GeoDataFrame(df.loc[m].to_frame().T)
        value=df.loc[m]['Índice de envelhecimento new']
        colorCode=mpl.colors.to_hex(cmap(value)[0:3])
        muni.plot(color=colorCode,ax=ax)
    plt.title(title)
    plt.show()
```

```python
printMap(df[df['Ano']==2010], azoresCountys,'Mapa dos Açores')
printMap(df[df['Ano']==2010], madeiraCountys,'Mapa dos Madeira')   
printMap(df[df['Ano']==2010], continentCountys,'Mapa do Continente')
```

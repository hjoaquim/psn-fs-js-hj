### Building a pseudo database with every county in Portugal
To do that we simply went to the Pordata website and ctrl+C / ctrl+V every county into a text file.
Some minor operations and Voilà!

Most of the times, simple tasks like that don't need to be automated, don't try to hard!! Just get the job done!

```python
# that must be our county database 
f = open(path_municipios, 'r', encoding="utf8")
lista_municipios = [m.strip() for m in f.readlines()]
lista_municipios
```

```python
df = df.append(pt_lines)
for mun in lista_municipios:
    d1 = pd.read_excel(os.path.join(path,'O seu municipio em numeros! ' + mun + '.xls'))
    d1 = d1[6:78]
    d1 = d1.drop(columns=['Unnamed: 2', 'Unnamed: 4', 'Unnamed: 5','Unnamed: 6'])
    d1.columns = ['Municipios'] + d1.loc[7][1:].tolist()
    d1.loc[6] = ['Ano', 2010, 2019]
    d1 = d1.drop(7)
    d1 = d1.transpose()
    d1 = d1.drop([d1.index[0]])
    #print(os.path.join(path,'O seu município em números! ' + mun + '.xls'))
    df = df.append(d1)
```

### Here we go! 
Right now our data cleasing is almost done; as we can see below there are just one more thing that catch the eye: to many missing values! So let's solve that and then we're ready to go!

```python
df
```

```python
df = df.dropna(axis='columns')
df
```
